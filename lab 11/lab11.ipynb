Схоже завдання насправді вирішується при аналізі реальних файлів у реальній антивірусній системі.

Для людини завдання здається очевидним, адже відразу видно, де чистий, а де шкідливий текст, але ось формалізувати різницю або тим більше пояснити її комп'ютеру - це значно складніше. Тут використовується машине навчання: спочатку системі потрібно дати зразки. Він на них "навчиться", а потім буде сам правильно відповідати, де що саме.


```python
import numpy as np
```

```python
ALPHABET = list("abcdefghijklmnopqrstuvwxyz")
len(ALPHABET)
```

```python
def build_bigram_counts(text: str, alphabet: str) -> np.ndarray:
    idx = {ch:i for i,ch in enumerate(alphabet)}
    L = len(alphabet)
    M = np.zeros((L, L))
    prev = None
    for ch in text:
        if ch not in idx:
            continue
        if prev is not None:
            M[idx[prev], idx[ch]] += 1.0
        prev = ch
    return M
```

```python
with open("texts/LOTR.txt", "r", encoding="utf-8") as f:
    train = f.read().lower()

train = "".join(ch for ch in train if ch in ALPHABET)

M_counts = build_bigram_counts(train, ALPHABET)
```

```python
M_counts
```

```python
def evaluation(text: str, alphabet: str, M: np.ndarray):
    idx = {ch:i for i,ch in enumerate(alphabet)}
    score = 1
    l = len(text)
    prev = None
    for ch in text:
        if ch not in idx:
            continue
        if prev is not None:
           score *= np.power(M[idx[prev], idx[ch]], 1/(l-1))
        prev = ch
    return score
```

```python
with open("texts/ch1Silmarillion.txt", "r", encoding="utf-8") as f:
    clean_text = f.read().lower()
print(len(clean_text))
evaluation(clean_text, ALPHABET, M_counts)
```

```python
def compute_threshold(clean_data, garbage_data, alphabet, M):
    clean_score = []
    garbage_score = []
    for i in range(len(clean_data)):
        clean_score.append(evaluation(clean_data[i], alphabet, M))
        garbage_score.append(evaluation(garbage_data[i], alphabet, M))
    
    clean_score = np.array(clean_score)
    garbage_score = np.array(garbage_score)
    
    return (clean_score.mean() + garbage_score.mean())/2
```

```python
compute_threshold(clean_data, garbage_data, ALPHABET, M_counts)
```
